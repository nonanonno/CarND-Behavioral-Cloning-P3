```python
import numpy as np
import sys
import matplotlib.pyplot as plt
import csv
from model import LoadTrainingData, SetupKeras, CreateModel
```

```python
X_train, y_train = LoadTrainingData(['datasets/track2/0/','datasets/track2/1/'], False)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(X_train[0])
plt.subplot(1,2,2)
plt.imshow(X_train[1])

print(y_train[0], y_train[1])
```

```python
SetupKeras()
model = CreateModel(X_train[0].shape)

val_loss = []
loss = []
```

```python
from sklearn.utils import shuffle
epochs = 20

for epoch in range(epochs):
    X_train, y_train = shuffle(X_train, y_train)
    stack = model.fit(X_train, y_train, validation_split=0.2,epochs=epoch+1, initial_epoch=epoch)
    val_loss += stack.history['val_loss']
    loss += stack.history['loss']
    
```

```python
x = range(len(loss))
plt.plot(x, val_loss, label="val_loss")

plt.plot(x, loss, label="loss")
plt.title("loss")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
